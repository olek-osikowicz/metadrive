```python
from utils.bayesian_optimisation import get_training_data, HDD_PATH, get_logger
from pathlib import Path
```

```python
rep_path = HDD_PATH / "searches" / "bayesopt_ucb" / "multifidelity" / "0"


train_df = get_training_data(rep_path=rep_path)
train_df
```

```python
logger = get_logger()
logger.setLevel("DEBUG")
target_fidelity = 60
```

```python
def need_to_verify_score(train_df):
    logger.debug("First check if we need to verify score")
    hf_train_df = train_df[train_df["fid.ads_fps"] == target_fidelity]
    current_best = hf_train_df["eval.driving_score"].min()

    last_dscore = train_df.iloc[-1]["eval.driving_score"]
    last_dscore
    logger.debug(f"{current_best = } {last_dscore = }")
    if last_dscore < current_best:
        logger.info("Last score is lower than current best, we need to verify!")
    else:
        logger.info("No need to verify search continues")


need_to_verify_score(train_df)
```

```python
logger.setLevel("INFO")
for batch in train_df.expanding(min_periods=5):
    need_to_verify_score(batch)

batch
```
